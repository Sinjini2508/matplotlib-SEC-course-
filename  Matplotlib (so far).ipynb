## Boxplot
A boxplot is a diagram which shows the distribution of values for a single variable




Box plot five number summary:
1. Minimum: The minimum value in the dataset
2. Median: The middle-most value in the dataset [50% of the values lie above the median, and 50% below]
3. First quartile: The middle-most value between the median and the minimum value [25% values lie below and 75 % above]
4. Maximum: The maximum value in the dataset
5. Third quartile: The middle-most value between median and the maximum value [75% of the values lie below and 25% above]


INTER-QUARTILE RANGE(The area between the first quartile and third quartile): Q3 - Q1


OUTLIERS:

A value 'X' is said to be an outlier if: 

X < Q1 - 1.5(IQR) or

X > Q3 + 1.5(IQR)

```python
plt.figure(figsize = (15, 10))

plt.boxplot(data['INT_SQFT'])

plt.xlabel('Boxplot', fontsize=20)
plt.ylabel('Square feet', fontsize=20)
```

CALCULATIONS 


```python
values = data['INT_SQFT']
maximum = max(values)
minimum = min(values)

Q1, Q3 = np.percentile(values, [25, 75])
med = np.median(values)

IQR = Q3-Q1


print("Minumum value: ", minimum)
print("First quartile: ", Q1)
print("Median: ", med)
print("Third quartile: ", Q3)
print("Maximum value: ", maximum)


```

```python
# To check whether an observation is an outlier or not

for x in data['INT_SQFT']:
    if (x > Q3 + 1.5*IQR) or (x < Q1 - 1.5*IQR):
        df = pd.DataFrame({'Observation': data['INT_SQFT'],
                            'Is it an outlier': 'Yes, an outlier'})
    else:
        df = pd.DataFrame({'Observation': data['INT_SQFT'],
                            'Is it an outlier': 'No, not an outlier'})
        
```
